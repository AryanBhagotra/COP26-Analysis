```python
import requests
import pandas as pd
import json
import datetime
import time
```

```python
pre_cop_start_date = datetime.datetime(2020, 12, 1)
pre_cop_end_date = datetime.datetime(2021, 10, 30)
pre_cop_start_date_timestamp = int(time.mktime(pre_cop_start_date.timetuple()))
pre_cop_end_date_timestamp = int(time.mktime(pre_cop_end_date.timetuple()))


during_cop_start_date = datetime.datetime(2021, 11, 1)
during_cop_end_date = datetime.datetime(2021, 11, 30)
during_cop_start_date_timestamp = int(time.mktime(during_cop_start_date.timetuple()))
during_cop_end_date_timestamp = int(time.mktime(during_cop_end_date.timetuple()))


post_cop_start_date = datetime.datetime(2021, 12, 1)
post_cop_end_date = datetime.datetime(2022, 1, 7)
post_cop_start_date_timestamp = int(time.mktime(post_cop_start_date.timetuple()))
post_cop_end_date_timestamp = int(time.mktime(post_cop_end_date.timetuple()))
```

```python
df = pd.read_csv('countries.csv')
df.head()
```

```python
apikey = '4a0c775b4f6cb0fc94614bd79be21d62'
```

```python
for index, row in df.iterrows():
    lat, lon = row['latitude'], row['longitude']
    country = row['COUNTRY']
    
    url = f'http://api.openweathermap.org/data/2.5/air_pollution/history?lon={lon}&lat={lat}&start={pre_cop_start_date_timestamp}&end={pre_cop_end_date_timestamp}&appid={apikey}'
    r = requests.get(url)
    data = r.json()
    with open(f'pre/{country}.json', 'w') as f:
        json.dump(data, f)
    time.sleep(1)
```
